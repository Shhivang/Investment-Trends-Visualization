```python
plt.figure(figsize = (15, 4), dpi = 150)

plt.subplot(1,3,1)
plt.hist(x = df['age'], edgecolor = 'white', color = '#936639')
plt.title('KDE Plot of Age column')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.subplot(1,3,2)
df['age'].plot(kind = 'kde', color = '#936639', title = 'KDE Plot of Age column')

plt.subplot(1,3,3)
sns.boxplot(data =df, x = df['age'], color = '#936639')
plt.title('Box Plot of Age column')
plt.xlabel('Age')

plt.tight_layout();
```

# 9. Data Visualization

```python
df.head()
```

```python
plt.figure(figsize = (18, 10), dpi = 150)

plt.suptitle('Detailed Analysis of Financial Behaviors and Preferences', fontsize = 20, fontweight = 'bold')

# Pie Chart: Investment Distribution by Gender
plt.subplot(2,3,1)
ax = df['gender'].value_counts().plot(
    kind = 'pie',
    startangle = 90,
    autopct = '%0.0f%%', 
    colors = ['#c2c5aa', '#936639'],
    wedgeprops = {'edgecolor' : 'white'}
)
plt.title('Investment Distribution by Gender', fontsize = 15, fontweight = 'bold')
plt.ylabel('')
ax.margins(y=0.1)

for i in ax.containers:
    ax.bar_label(i, fmt = '{:,.0f}')
    plt.xticks(rotation = 45)

# Horizontal Bar Chart: Stock Market Preferences by Gender
plt.subplot(2,3,2)
ax = sns.countplot(
    data = df,
    y= 'Stock_Marktet',
    hue = 'gender',
    palette = [ '#936639', '#c2c5aa'],
    edgecolor = 'white'
)
plt.xlabel('Frequency')
plt.title('Stock Market Preferences by Gender', fontsize = 15, fontweight = 'bold')
ax.margins(x = 0.15)

for i in ax.containers:
    ax.bar_label(i, fmt = '{:,.0f}')

plt.legend(loc = (1.05, 0.8))


# Bar Chart: Investment Distribution by Avenue
plt.subplot(2,3,3)
ax = sns.countplot(
    data = df,
    x = 'Avenue',
    hue = 'gender',
    palette = [ '#936639', '#c2c5aa'],
    edgecolor = 'white'
)
ax.margins(y = 0.15)
plt.title('Investment Distribution by Avenue', fontsize = 15, fontweight = 'bold')

for i in ax.containers:
    ax.bar_label(i, fmt = '{:,.0f}')
    plt.xticks(rotation = 45)

plt.legend(loc = (1.05, 0.8))


# Bar Chart: Age-wise Distribution of Investment
plt.subplot(2,3,4)
ax = sns.countplot(
    data = df,
    x = 'age',
    color = '#c2c5aa'
)
ax.margins(y=0.15)

for i in ax.containers:
    ax.bar_label(i, fmt = '{:,.0f}')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age-wise Distribution of Investment', fontsize = 15, fontweight = 'bold')


# Donut Chart: Distribution of Investment Tenures
plt.subplot(2,3,5)
plt.title('Distribution of Investment Tenures', fontsize = 15, fontweight = 'bold')
ax = df.groupby('Duration')['gender'].count().plot(
    kind = 'pie',
    autopct = '%0.0f%%',
    startangle = 45,
    colors = ['#c2c5aa', '#7f4f24', '#936639', '#414833'],
    wedgeprops = {'edgecolor': 'white'},
    fontsize = 9
)

center_circle = plt.Circle((0, 0), 0.75, color='white', fc='white', linewidth=0)
ax.add_artist(center_circle)

plt.ylabel('')


# Pie chart: Break-down of Expected Returns
plt.subplot(2,3,6)
df.groupby('Expect')['age'].count().plot(
    kind = 'pie',
    colors = ['#7f4f24', '#c2c5aa', '#936639'],
    wedgeprops = {'edgecolor': 'white'},
    autopct = '%0.0f%%',
    startangle = 45,
    fontsize = 9
)
plt.ylabel('')
plt.title('Break-down of Expected Returns', fontsize = 15, fontweight = 'bold')


plt.subplots_adjust(top=0.85,  # Reserve space at the top for the super title
                    wspace=0.4,  # Width space between subplots
                    hspace=0.6)  # Height space between subplots


plt.tight_layout(rect=[0, 0, 1, 0.95])

plt.savefig('Investment_Trends_Visualization.png');
```
